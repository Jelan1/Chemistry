####  Translating from SMILES to Molecule Object

```python
hydroquinone_smiles = 'c1cc(ccc1O)O' # Hydroquinone raw smiles code

molecule = Chem.MolFromSmiles(hydroquinone_smiles) # Hydroquinone object in database


molecular_weight = Descriptors.MolWt(molecule) # Many Descriptors available, I just chose MolWt (weight)
print(f'Molecular Weight: {molecular_weight}') # ignore

molecule # printing the Hydroquinone object
```

## A Closer Look at Some Basic Functions

```python
# Setting up, same as last step but for more molecules now. The specifics of the code are not important right now

hydroquinone = 'c1cc(ccc1O)O hydroquinone'  
phenylalanine = 'C1=C(NC=N1)CC(C(=O)O)N phenylalanine '
histidine = 'NC(C(=O)O)Cc1ccccc1 histidine'
cysteine = 'C(C(C(=O)O)N)S cysteine '


list_of_smiles = [hydroquinone,phenylalanine,histidine,cysteine]
mol_list = [Chem.MolFromSmiles(molecule) for molecule in list_of_smiles]
image = Draw.MolsToGridImage(mol_list,molsPerRow=4)
image
# We have 4 molecules to invevstigate
```

### Filtering A List of Molecules Based on Substructure 

 
#### - Given the 4 molecules above, return only the molecule(s) containing a given substructure.

#### - Additionally, Place a Red Highlight on the  queried substructure when displaying the molecule's 2-D image 

##### Example 1

```python
### FIND MOLECULES CONTAINING THE BELOW SUBSTRUCTURE
Chem.MolFromSmiles('C(=O)')
```

```python
# ! The Code to carry this out is messy. Please Ignore and focus on the results 

substructure = 'C(=O)'
substructure = Chem.MolFromSmiles(substructure)  
lis = [[mol,mol.GetProp('_Name')] for mol in mol_list if (mol.HasSubstructMatch(substructure) is True)]
for i,y in enumerate(lis):
    m = lis[i][0]
    print(lis[i][1])
    m.GetSubstructMatches(substructure)
    display(m)
```
